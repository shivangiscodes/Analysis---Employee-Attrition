```python
import pandas as pd 

df=pd.read_csv('C:\\Users\\Shivam\\Desktop\\Trimester-5\\Big Data Analytics\\CAC1\\HRdata.csv')

df.head()
```

```python
# 1. Handle missing values
# Fill missing 'gender' with 'Unknown'
df['gender'] = df['gender'].fillna('Unknown')
```

```python
# Fill missing 'enrolled_university' with 'no_enrollment'
df['enrolled_university'] = df['enrolled_university'].fillna('no_enrollment')
```

```python
# Fill missing 'major_discipline' with 'Other'
df['major_discipline'] = df['major_discipline'].fillna('Other')
```

```python
# Fill missing 'company_size' and 'company_type' with 'Unknown'
df['company_size'] = df['company_size'].fillna('Unknown')
df['company_type'] = df['company_type'].fillna('Unknown')
```

```python
# Save cleaned data to a separate CSV file
cleaned_file_path = "HRdata(PowerBI).csv"  # Specify your desired file path and name
df.to_csv(cleaned_file_path, index=False)  # Use index=False to exclude the index column
print(f"Cleaned data saved to {cleaned_file_path}")
```

```python
# 2. Normalize values
# Convert experience to numeric ranges
def convert_experience(exp):
    if exp == '>20':
        return 21
    elif exp == '<1':
        return 0
    else:
        return int(exp)

df['experience'] = df['experience'].apply(convert_experience)

# Normalize 'last_new_job' values
def convert_last_new_job(job):
    if job == '>4':
        return 5
    elif job == 'never':
        return 0
    else:
        return int(job)

df['last_new_job'] = df['last_new_job'].apply(convert_last_new_job)
```
